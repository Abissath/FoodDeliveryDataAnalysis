# Food Delivery Demand Analysis in Ghana

 

### Context

In major cities in Ghana such as Accra , Kumasi, and Takroadi, the number of restaurants and fast-food outlets has been increasing rapidly. Many students, office workers, and busy professionals rely on these restaurants due to their demanding schedules and limited time to cook.

Online food delivery services have therefore become increasingly popular. These platforms allow customers to conveniently order meals from their favorite restaurants using a smartphone application and have them delivered directly to their homes or workplaces.

In Ghana, several digital food delivery platforms connect customers with restaurants through a single mobile application. When a customer places an order through the app, the restaurant receives the request and begins preparing the food once the order is confirmed. A delivery rider is then assigned by the platform to pick up the order from the restaurant.

Using GPS navigation, the delivery rider travels to the restaurant, collects the food package once it is ready, and confirms the pickup in the mobile app. The rider then proceeds to the customer’s location to deliver the order. After the food is delivered, the rider confirms the drop-off in the app, and the customer can rate the service and food quality.

The food delivery platform generates revenue by charging restaurants a fixed commission or margin on each order processed through the system.

### Objective
The food delivery platform has collected a large dataset of orders placed by registered customers through its online platform in Ghana.

**The company wants to analyze this data to better understand:**

**As a Data Scientist working with this food delivery company, my task is to analyze the order data and generate insights that will help the company:**

### Data Dictionary

* order_id
Unique identification number assigned to each food order.

* customer_id
Unique ID representing the customer who placed the order.

* restaurant_name
Name of the restaurant where the order was placed.

* cuisine_type
Type of cuisine ordered by the customer (e.g., Ghanaian, Chinese, Italian, Fast Food, etc.).

* cost
Total cost of the food order in Ghanaian Cedis (GHS).

* day_of_the_week
Indicates whether the order was placed on a weekday (Monday–Friday) or weekend (Saturday–Sunday).

* rating
Customer rating for the order on a scale from 1 to 5, where 5 represents the highest satisfaction.

* food_preparation_time
Time (in minutes) taken by the restaurant to prepare the food.
This is calculated as the difference between the restaurant order confirmation time and the delivery rider pickup confirmation time.

* delivery_time
Time (in minutes) taken by the delivery rider to deliver the food to the customer.
This is calculated as the difference between the pickup confirmation time and the order delivery confirmation time.
