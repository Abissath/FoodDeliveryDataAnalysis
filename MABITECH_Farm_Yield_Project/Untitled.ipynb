## Importing libraries for the project

```python
import pandas as pd
import numpy as np
np.random.seed(42)
data = pd.DataFrame({"rainfall":np.random.randint(700, 1500, 100),"temperature":np.random.randint(24,32,100),})
# Create yield based on a simple relationship
data["yield"] = (data["rainfall"]*0.002) + (data["temperature"]* 0.1)+ np.random.normal(0,0.3,100)
print(data.head())
```

```python
data.shape
```

## Importing Algorithm for the project


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
X = data[['rainfall', 'temperature']]
y = data['yield']
#Splitting the data for training and testing and the test size  is 20 percent
X_train , X_test, y_train,y_test = train_test_split(X, y , test_size=0.2)
```

```python
#Training the model with random forest regressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
# Making a Prediction with the model
yiel_pred = model.predict([[1200, 28]])
# train_test_score = model.score(X_train, y_train)
# test_test_score = model.score(X_test, y_train)
# print("Predicting for train test score:{}".format(train_test_score))
print("Yield prediction when when we havest rainfall of 1200mm and the temperature is 28⁰C: ",yiel_pred[0]);

```
